# Inferential Statistics

Let's see if there is any correlation between provided sales data in the training set, and the clean weather data. To start, I will clean the weather data and remove any weather observations with missing data. Also, I will merge the training sales data with the weather data by way of the key CSV provided, linking stores with their respective station. A dictionary will be created to house each store's dataset. From here I will measure the correlation between units sold and general precipitation. To supplement I will also measure the correlation between units sold and significant rain (> 1 inch), and the correlation between units sold and significant snowfall (> 1 inch).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

raw_weather = pd.read_csv('data/weather.csv')
sales_train = pd.read_csv('data/train.csv')
key = pd.read_csv('data/key.csv')

weather_df = raw_weather[['date','station_nbr','tavg','preciptotal']]
replacements = {'tavg' : {r'M' : np.nan},
                'preciptotal' : {r'T' : 0.00, r'M' : np.nan}}
clean_weather = weather_df.replace(replacements, regex=True)
clean_weather.dropna(inplace=True)
clean_weather['tavg'] = clean_weather['tavg'].apply(pd.to_numeric)
clean_weather['preciptotal'] = clean_weather['preciptotal'].apply(pd.to_numeric)

sales_station = pd.merge(sales_train, key, left_on='store_nbr', right_on='store_nbr')
sales_weather = pd.merge(sales_station, clean_weather, left_on=['date','station_nbr'], right_on=['date','station_nbr'])

store_sales_weather = {}
for x in range(1,45):
    store_sales_weather['s{0}'.format(x)] = sales_weather.loc[sales_weather.store_nbr == x]
    
store_sales_weather['s1']
```

Looking at the scatterplot of units sold vs general precipitation for store 19, you can notice the majority of observations clustered around low precipitation counts. However, when filtering by significant rain (anything above 1 inch), there seems to be a very slight positive relationship; visually, the observations still do not indicate a strong correlation, and there is a high concentration of observations with zero units sold no matter the rain levels. The same can also be said for the significant snow filtered plot, with most observations having zero units sold. Now I will calculate the correlations between units sold and each of the weather categories, for each store.

```python
x = 0
for store in d.keys():
    x+=1
    big_rain = d[store][(d[store]['preciptotal']>1.00) & (d[store]['tavg']>32)]
    big_snow = d[store][(d[store]['preciptotal']>1.00) & (d[store]['tavg']<=32)]
    print('The correlation coefficients for store %d:' % x)
    general = d[store]['units'].corr(d[store]['preciptotal'])
    rain = big_rain['units'].corr(big_rain['preciptotal'])
    snow = big_snow['units'].corr(big_snow['preciptotal'])
    print('General correlation between units sold and precipitation: %f' % general)
    print('Correlation between units sold and significant rain (over 1 inch): %f' % rain)
    print('Correlation between units sold and significant snow (over 1 inch): %f' % snow)
    print('****************************************************************************')
    print()
```
